```python
import pandas as pd
dataframe=pd.read_table("chicago.txt",sep="\t",header=None, names=["shop_id","shop_name","tags"])
dataframe
```

```python
list_tags=list(dataframe["tags"])
list_tags
```

```python
list_tag1=[]
for elem in list_tags:
    list_tags_split=elem.split(' ')#这里用list_tags_split=elem.split('\s')结果无法分开，为什么？
    list_tag1=list_tag1+list_tags_split
print(len(list_tag1))
list_tag1=list(set(list_tag1))
list_tag1.sort()
print(len(list_tag1))
list_tag1
```

```python
list_tag2=[]
for elem in list_tags:
    list_temp=elem.split(' ')
    list_tag2.append(list_temp)
print(len(list_tag2))
list_tag2
```

```python
list_tag3=[]
for elem1 in list_tag2:
    list_temp=[]
    for elem2 in elem1:
        list_temp.append(int(elem2))
    list_tag3.append(list_temp)
list_tag3
```

```python
#构造 所有餐厅*所有标签 的01矩阵
#使用sparse.dok_matrix()构造非稀疏矩阵
from scipy import sparse
import numpy as np
S = sparse.dok_matrix((676, 254), dtype=np.int8)#构造

for i in range(676):#往里面插入值（若不插入值则刚构造完的非稀疏矩阵里面全是0）
    for j in range(254):
        for elem in list_tag3[i]:
            S[i,elem]=1;#注意!!!给稀疏矩阵中元素赋值时格式为S[i,elem]=1，而不是S[i][elem]=1
S.toarray()
```

```python
array_test=S.toarray()
array_test[0][214]#验证一下构造的稀疏矩阵是否正确
```

```python
def diancheng(list1,list2):
    result=0
    for i in range(len(list1)):
        result=result+list1[i]*list2[i]
    return result
```

```python
from scipy import sparse
import numpy as np

#构造 所有餐厅*其他餐厅和它的相似度 的矩阵（矩阵值为相似度，越大越相似）
def get_similaritymatrix_from_labelmatrix(labelmatrix):
    len_labelmatrix=len(labelmatrix)
    S = sparse.dok_matrix((len_labelmatrix, len_labelmatrix), dtype=np.int8)#构造
    for i in range(len_labelmatrix):
        for j in range(i+1,len_labelmatrix):
            tmp=diancheng(labelmatrix[i],labelmatrix[j])
            S[i,j]=tmp
            S[j,i]=tmp
    return S
array_test2=get_similaritymatrix_from_labelmatrix(array_test).toarray()
array_test2
```

```python
#构造 所有餐厅*与之相似度最高的n个餐厅 的非01矩阵（矩阵值为餐厅id）

n=10

def compare_dict(dict_elem):
    return dict_elem[1]

def chose_bestn_from_similaritymatrix(similaritymatrix):
    result=[]
    #对矩阵中的每一行（每一行就是每一个餐厅与其他所有餐厅的相似度的list）
    for elem in similaritymatrix:
        # 将list构造成dict
        dict_temp={}
        for i in range(len(elem)):
            dict_temp[str(i)]=elem[i]
        
        # 对dict根据value排序（降序）
        list_temp_sorted=sorted(dict_temp.items(),key=compare_dict,reverse=True)#reverse=True降序
        
        # 选出排好序的前两名
        list_temp_sorted_topn=[]
        for i in range(n):#选前n名
            list_temp_sorted_topn.append(int(list_temp_sorted[i][0]))
        
        # 将其插入 所有餐厅*与之相似度最高的n个餐厅 的非01矩阵中
        result.append(list_temp_sorted_topn)
    return result

recommend_list=chose_bestn_from_similaritymatrix(array_test2)
recommend_list
```
