```python
def diancheng(list1,list2):
    result=0
    for i in range(len(list1)):
        result=result+list1[i]*list2[i]
    return result
```

```python
from scipy import sparse
import numpy as np

#æž„é€  æ‰€æœ‰é¤åŽ…*å…¶ä»–é¤åŽ…å’Œå®ƒçš„ç›¸ä¼¼åº¦ çš„çŸ©é˜µï¼ˆçŸ©é˜µå€¼ä¸ºç›¸ä¼¼åº¦ï¼Œè¶Šå¤§è¶Šç›¸ä¼¼ï¼‰
def get_similaritymatrix_from_labelmatrix(labelmatrix):
    len_labelmatrix=len(labelmatrix)
    S = sparse.dok_matrix((len_labelmatrix, len_labelmatrix), dtype=np.int8)#æž„é€ 
    for i in range(len_labelmatrix):
        for j in range(i+1,len_labelmatrix):
            tmp=diancheng(labelmatrix[i],labelmatrix[j])
            S[i,j]=tmp
            S[j,i]=tmp
    return S
array_test2=get_similaritymatrix_from_labelmatrix(array_test).toarray()
array_test2
```

```python
#æž„é€  æ‰€æœ‰é¤åŽ…*ä¸Žä¹‹ç›¸ä¼¼åº¦æœ€é«˜çš„nä¸ªé¤åŽ… çš„éž01çŸ©é˜µï¼ˆçŸ©é˜µå€¼ä¸ºé¤åŽ…idï¼‰

n=10

def compare_dict(dict_elem):
    return dict_elem[1]

def chose_bestn_from_similaritymatrix(similaritymatrix):
    result=[]
    #å¯¹çŸ©é˜µä¸­çš„æ¯ä¸€è¡Œï¼ˆæ¯ä¸€è¡Œå°±æ˜¯æ¯ä¸€ä¸ªé¤åŽ…ä¸Žå…¶ä»–æ‰€æœ‰é¤åŽ…çš„ç›¸ä¼¼åº¦çš„listï¼‰
    for elem in similaritymatrix:
        # å°†listæž„é€ æˆdict
        dict_temp={}
        for i in range(len(elem)):
            dict_temp[str(i)]=elem[i]
        
        # å¯¹dictæ ¹æ®valueæŽ’åºï¼ˆé™åºï¼‰
        list_temp_sorted=sorted(dict_temp.items(),key=compare_dict,reverse=True)#reverse=Trueé™åº
        
        # é€‰å‡ºæŽ’å¥½åºçš„å‰ä¸¤å
        list_temp_sorted_topn=[]
        for i in range(n):#é€‰å‰nå
            list_temp_sorted_topn.append(int(list_temp_sorted[i][0]))
        
        # å°†å…¶æ’å…¥ æ‰€æœ‰é¤åŽ…*ä¸Žä¹‹ç›¸ä¼¼åº¦æœ€é«˜çš„nä¸ªé¤åŽ… çš„éž01çŸ©é˜µä¸­
        result.append(list_temp_sorted_topn)
    return result

recommend_list=chose_bestn_from_similaritymatrix(array_test2)
recommend_list
```
